```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read data from CSV as DataFrame
data = pd.read_csv("./papi_reports/data.csv", delimiter=',')
```

```python
# Plot Data
plt.figure()
data.plot(x='Resolution', y=data.columns[1:])
plt.xlabel('Resolution')
plt.ylabel('Accesses and Misses')
plt.savefig('./papi_reports/Graph_1')

```
